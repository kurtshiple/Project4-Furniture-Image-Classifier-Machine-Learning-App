#### Loading and Evaluating the loaded model

```python
# Load the model

import os
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
from PIL import Image 
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing import image
```

```python
# def collect_info():
#     pass
```

```python
model = load_model("model1.h5")
```

```python
# Define default image size for our model
image_size = (100, 100)
```

```python
# Load the image and resize to default image size
image_path = os.path.join(".", "test.png") # 4933.jpeg
img = image.load_img(image_path, target_size=image_size)
plt.imshow(img)
```

```python
# Preprocess image for model prediction
# This step handles scaling and normalization for VGG19
x = image.img_to_array(img)
x = np.expand_dims(x, axis=0)
# x = preprocess_input(x)
x_ques = x.astype("float32")/255.0
x_quest = x_ques.reshape(x_ques.shape[0], x_ques.shape[1],x_ques.shape[2],3)
```

```python
x_quest.shape
```

```python
model.predict(x_quest).round()
```

```python
print(f"Predicted class: {model.predict_classes(x_quest)}")
```
